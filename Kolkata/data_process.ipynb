```python
import matplotlib.pyplot as plt
import csv
```

```python
def parameter_process(list_param, name, date_param):
    plt.stem(date_param, list_param)
    plt.xlabel('Date-Time')
    plt.xticks(fontsize=3.5)
    plt.ylabel(name)
    plt.show()
    # plt.figure(figsize=(110, 8))

# def parameter_final(name):


def process_csv(filename):
    data=[]
    time=[]
    co=[]
    o3=[]
    with open(filename) as file:
        reader=csv.reader(file)
        for row in reader:
            data.append(row)
            if row==[]:
                data.remove(row)
    data=data[1:]
    for i in data:
        o3.append(i[4])
        time.append(i[-1])
        co.append(i[0])
    parameter_process(co, 'Carbon-Monoxide', time)
    parameter_process(o3, 'Ozone', time)
    a=plt.stackplot(time, co, o3, labels=['Carbon-Monoxide', 'Ozone'])
    plt.show(a)
```
